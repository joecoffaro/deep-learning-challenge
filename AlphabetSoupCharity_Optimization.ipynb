## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial columns 'EIN'
application_df = application_df.drop(columns=['EIN'])
application_df
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at NAME value counts to identify and replace with "Other"
name_counts = application_df['NAME'].value_counts()
name_counts[name_counts >= 10]
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_val = 9
names_to_replace = name_counts[name_counts <= cutoff_val].index.tolist()

# Replace in dataframe
for name in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(name,"Other")

# Check to make sure replacement was successful
application_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
app_counts = application_df['APPLICATION_TYPE'].value_counts()
app_counts
```

```python
# You may find it helpful to look at APPLICATION_TYPE value counts > 100
app2_counts = app_counts[app_counts > 100]
app2_counts
```

```python
# Choose a cutoff value and create a list of APPLICATION_TYPE to be replaced
# use the variable name `apps_to_replace`
cutoff_value = 500
apps_to_replace = app_counts[app_counts < cutoff_value].index.tolist()

# Replace in dataframe
for app in apps_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_modified = pd.get_dummies(application_df)
application_df_modified
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_modified.drop(columns=['IS_SUCCESSFUL']).values
y = application_df_modified['IS_SUCCESSFUL'].values


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
no_input_features = len(X_train[0])
hidden_node1 = 120
hidden_node2 = 50
hidden_node3 = 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_node1, activation="relu", input_dim=no_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_node2, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_node3, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
